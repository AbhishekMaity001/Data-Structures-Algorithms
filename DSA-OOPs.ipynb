# Abstraction --> (interface + implementation)
- To hide the actual implementation
- the idea behind this is , we deal with interface without worrying about implementation
- by default in python abstraction is not implemented, there are libraries to implement that. abstract method and abstract classes
- Abstract method ---> is a method that does not have implementation (not a concrete method)

```python
from abc import abstractmethod, ABC
```

    This code will define what is the criteria which will define the payment method. This will actually enforce that all the payment 
    methods should have a mandatory pay and cancel methods inside it. If it does not have these methods then python will not allow only to create a class
    Mandatory method overriding for all abstract method

```python
class PaymentMethod(ABC):
    """
    This is the Interface
    """
    @abstractmethod
    def pay(self):
        pass

    @abstractmethod
    def cancel(self):
        pass
```

```python
class CreditCardPayment(PaymentMethod):
    """
    This is the implementation of the Interface
    """
    
    def pay(self):
        print("Transcation completed successfully...")
    def cancel(self):
        print("Transcation cancelled...")

class UPIPayment(PaymentMethod):
    """
    This is the implementation of the Interface
    """
    
    def pay(self):
        print("Transcation completed successfully...")
    def cancel(self):
        print("Transcation cancelled...")

class InternetBankingPayment(PaymentMethod):
    """
    This is the implementation of the Interface
    """
    
    # def pay(self):
    #     print("Transcation completed successfully...")
    # def cancel(self):
    #     print("Transcation cancelled...")
        
    def make_payment(self):
        print("Transcation completed successfully...")
    def cancel_payment(self):
        print("Transcation cancelled...")
```
