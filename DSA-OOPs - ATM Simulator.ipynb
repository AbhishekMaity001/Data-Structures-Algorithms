# OOPs mini Project - ATM Simulator

```python
class BankAccount:
    # class variable
    bank_acc_name = "Axis Bank"
    
    def __init__(self, account_number:str, pin:int, balance:float):
        self.account_number = account_number
        self.pin = pin
        self.balance = balance

    def check_balance(self):
        return self.balance

    def deposit(self, amount:float):
        if amount<0:
            raise ValueError("Amount deposited cant be in negative!")
        else:
            self.balance += amount
            print(f"Amount deposited Rs {amount} successfully!, Your new balance is : Rs {self.balance}")

    def withdraw(self, amount:float):
        if amount>0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Amount withdrawn Rs {amount}, Your new balance is : Rs {self.balance}")
            else:
                raise ValueError(f"Amount withdraw is greater than available balance Rs {self.balance}!")
            
        else:
            raise ValueError("Amount withdraw cant be in negative!")

        
```

```python
class ATM:
    """
    ATM class will register an account, authenticate and then allow the user to deposit and withdraw money from the account.
    """
    def __init__(self):
        self.registered_accounts = {}  # we are storing {acc number : acc object}

    def register_account(self, account:BankAccount):
        self.registered_accounts[account.account_number] = account

    def authenticate(self, account_number, pin):
        account = self.registered_accounts.get(account_number, None)
        if account is None:
            raise ValueError("INVALID ACCOUNT NUMBER!!")
        if pin == account.pin:
            print("Login Successfull...")
            return account
        else:
            raise ValueError("INVALID PIN!")

    @staticmethod
    def display_options():
        print("""
                Select the correct option:
                1. check your balance
                2. withdraw
                3. deposit
                4. Exit
        """)





    
```

    Next Steps...

    1. Create the bank accounts
    2. Register the bank account with the atm
    3. Ask the user to enter the acc number
    4. Enter the pin
    5. Authenticate and then show options
    6. As per option selected perform deposit or withdraw

```python
atm = ATM()

atm.register_account(BankAccount("1", 1234, 10000))
atm.register_account(BankAccount("2", 5678, 20000))


while True:
    account_number = input("\n Enter the account number:")
    if account_number not in atm.registered_accounts:
        print(f"This account cant be serviced. Please try with some other account...")
        continue
    
    pin = int(input("Please Enter the pin: "))
    
    account = atm.authenticate(account_number=account_number, pin=pin)
    
    while True:
        atm.display_options()
        choice = int(input("\n Enter your choice::"))
        
        
        if choice==1:
            balance = account.check_balance()
            print(f"Your account balance is Rs {balance}")
        elif choice==2:
            withdraw_amount = int(input("Enter the Amount you want to withdraw :: "))
            account.withdraw(withdraw_amount)
        elif choice==3:
            deposit_amount = int(input("Enter the Amount you want to deposit :: "))
            account.deposit(deposit_amount)
        else:
            print("Thankyou for banking with us! Have a nice day!")
            break
```
