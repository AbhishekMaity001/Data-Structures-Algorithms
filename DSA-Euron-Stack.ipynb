# Stack

## ADT of Stack
- push
- pop
- seek/top
- size
- is_empty
- is_full

```python
class ListStack:
    """
    Stack implementation using a list 
    Under the hood we will be doing list operations only but interface will be of a stack
    Push, pop, seek, size
    """
    def __init__(self):
        self._list = []

    def push(self, value):
        """
        Push a value in a stack.
        TC: O(1)
        """
        self._list.append(value)

    def is_empty(self):
        return self.size()==0

    def pop(self):
        """
        Remove a value from top
        TC: O(1)
        """
        # check if stack is empty
        if self.is_empty():
            raise Exception("Stack is empty")

        return self._list.pop()

    def top(self):
        """
        only return the value
        """
        if self.is_empty():
            raise Exception("Stack is empty")
        return self._list[-1]

    def size(self):
        """
        
        """
        return len(self._list)

    def to_list(self):
        return list(self._list)  # creating a copy
        
    def __str__(self):
        return ("\n"+"__"+"\n").join([str(x) for x in self._list])

    def __repr__(self):
        return ("\n"+"---"+"\n").join([str(x) for x in reversed(self._list)])
```

# Implementation of Stack using LinkedList

- We will treat the head of the Linked list at the top of the stack

- Stack Operations
1. push : add the new value to the head of the linked list O(1)
2. pop : remove the head of the linked list O(1)
3. seek : return the value of head O(1)

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
class LinkedList:
    """
    Implementation of linkedlist using stack
    """
    def __init__(self):
        self._head = None  # indication that user should not use this variable
        self._size = 0

    def __repr__(self):
        list_of_elements = self._traverse()
        return "-->".join([str(x) for x in list_of_elements])

    def _traverse(self):
        output = []
        if self.is_empty():
            print("LINKED LIST IS EMPTY!")
            return output
        else:
            curr = self._head
            while curr:
                output.append(curr.value)
                curr = curr.next
            return output
        
    def is_empty(self):
        return self._size == 0

    def push(self, value):
        """
        Add a new value to the linked list => TC : O(1)
        """
        newNode = Node(value)

        # edge cases to check first if the Linked list is empty or not
        if self.is_empty():
            self._head = newNode
        else:
            # if not empty then just shift the head to the new node
            curr = self._head
            self._head = newNode
            self._head.next = curr
        self._size += 1
        
        return newNode.value
    def pop(self):
        """
        delete a value from the top
        """
        # edge cases to check first if the Linked list is empty or not
        if self.is_empty():
            return f"LINKED LIST IS ALREADY EMPTY!"
        else:
            curr = self._head
            self._head = self._head.next
            return curr.value

    def top(self):
        if self.is_empty():
            return f"LINKED LIST IS ALREADY EMPTY!"
        else:
            return self._head.value
        

node1 = LinkedList()  
```
