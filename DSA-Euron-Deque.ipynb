# Implementation of Stack using a Deque

- Deque --> Double ended queue
- variety of a doubly linked list

- Speciality of a doubly linked list
- insert at beginning O(1)
- insert at end
- remove from beginning
- remove from end

```python
from collections import deque
```

```python
my_deque = deque()
```

```python
my_deque
```

```python
my_deque.append(55)
my_deque
```

```python
my_deque.appendleft(77)
my_deque.appendleft(88)
my_deque.appendleft(98)
my_deque
```

```python
my_deque.pop()
```

```python
my_deque.popleft()
```

```python
my_deque
```

### Deque
- Deques are a generalization of stacks and queues. because ------> in stack you will remove and add element from same end and in
Queue also you will remove from one end and add element from another end
- all the operations in Deque happen in O(1) time
