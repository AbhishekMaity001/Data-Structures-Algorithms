### Common operations in LinkedList
- append
- create
- delete
- search

```python
## Creating a class Node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


## Create a Linked List class
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0
        
    def is_empty(self):
        return self.size == 0
        
    def append(self, value):
        """
        case 1 : if linkedlist is empty then assign new node as head
        case 2 : if linkedlist is having data then traverse to tail and set new node after tail.
        """
        new_node = Node(value)
        if self.is_empty():
            self.head = new_node
        else:
            curr = self.head

            # traversing till current is None
            while curr.next:
                curr = curr.next
            curr.next = new_node
        self.size += 1

    def __repr__(self):
        list_of_elements = self._traverse()
        return " -> ".join([str(x) for x in list_of_elements]) + " -> None"
    
    def _traverse(self):
        elements = []
        if self.is_empty():
            print("list is empty....")
            return elements
            
        curr = self.head
        while curr:
            elements.append(curr.data)
            curr = curr.next
        return elements
        
        
        
```
